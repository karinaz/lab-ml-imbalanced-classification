```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

def evaluate_model(model, X_train, X_test, y_train, y_test):
    # Predictions on train set
    train_pred = model.predict(X_train)
    train_accuracy = accuracy_score(y_train, train_pred)
    train_precision = precision_score(y_train, train_pred, pos_label='Yes')
    train_recall = recall_score(y_train, train_pred, pos_label='Yes')

    # Predictions on test set
    test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(y_test, test_pred)
    test_precision = precision_score(y_test, test_pred, pos_label='Yes')
    test_recall = recall_score(y_test, test_pred, pos_label='Yes')

    return train_accuracy, train_precision, train_recall, test_accuracy, test_precision, test_recall

# Evaluating Logistic Regression Model
logistic_regression_metrics = evaluate_model(log_model, X_train, X_test, y_train, y_test)

# Evaluating KNN Classifier
knn_classifier_metrics = evaluate_model(knn_classifier, X_train, X_test, y_train, y_test)

# Evaluating Decision Tree Classifier
decision_tree_metrics = evaluate_model(decision_tree_classifier, X_train, X_test, y_train, y_test)
```

```python
def print_results(model_name, metrics):
    print(f"Results for {model_name}:")
    print(f"  Training Set - Accuracy: {metrics[0]:.4f}, Precision: {metrics[1]:.4f}, Recall: {metrics[2]:.4f}")
    print(f"  Testing Set - Accuracy: {metrics[3]:.4f}, Precision: {metrics[4]:.4f}, Recall: {metrics[5]:.4f}\n")
    
print_results("Logistic Regression", logistic_regression_metrics)
print_results("KNN Classifier", knn_classifier_metrics)
print_results("Decision Tree Classifier", decision_tree_metrics)
```
